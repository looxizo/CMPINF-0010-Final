First we must import needed libaries and initialize dictionaries.

```python
import pandas as pd
import csv
from matplotlib import pyplot as plt

# load data
with open("speeds.csv") as f:
    dataset = [row for row in csv.reader(f)]
neighSpeeds = {}
neighCount = {}
neighAveSpeed = {}
```

Then, by looping through the dataset, I add each entry's neighborhoods and recorded median speed to their respective dictionary. The if-else statements seperate the entries into ones that are already in the dictionary and those who are not. NeighSpeeds keep track of all the median speeds added together and the neighCount keeps count of how many speeds there, so we can find the average of the median speeds in every entry. 

```python
for row in dataset[1:]:
    if row[11] != '':
        if row[17] in neighSpeeds:
            neighSpeeds[row[17]] = float(neighSpeeds[row[17]]) + float(row[11])
            neighCount[row[17]] = int(neighCount[row[17]])+1
        else:
            neighSpeeds[row[17]] = float(row[11])
            neighCount[row[17]] = 1

keys = list(neighSpeeds.keys())
for i in range(len(neighSpeeds)):
    key = keys[i]
    neighAveSpeed[key] = float(neighSpeeds[key])/float(neighCount[key])
```

Now looks look at the top average speeds

```python
neighAveSpeed_series = pd.Series(neighAveSpeed, name = "Speed")
neighAveSpeed_series.sort_values(ascending=False)[:20].plot.bar()
plt.title('Average Speed per Neighborhood')
plt.xlabel('Neighborhoods')
plt.ylabel('Speed (mph)')
```
